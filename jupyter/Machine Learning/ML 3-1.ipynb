````python
import pandas as pd
import numpy as np

from sklearn import preprocessing
````

````python
NBA_FILE_PATH = "data/NBA_player_of_the_week.csv"
nba_player_df = pd.read_csv(NBA_FILE_PATH)
````

````python
nba_player_df.head()
````

````python
nba_player_df.describe()
````

````python
height_weight_age_df = nba_player_df[['Height CM', 'Weight KG', 'Age' ]]
height_weight_age_df.head()
````

### min-max scaler

````python
#min-max scaler 만들어 주는 도구
scaler = preprocessing.MinMaxScaler()
normalized_data = scaler.fit_transform(height_weight_age_df)
normalized_data
````

````python
normalized_df = pd.DataFrame(normalized_data, columns = ['Height', 'Weight', 'Age'])
normalized_df
````

````python
normalized_df.describe()
````

### standardization

```standard_scaler = preprocessing.StandardScaler()```

MinMaxScaler 대신 StandardScaler import

````python
standard_scaler = preprocessing.StandardScaler()
````

````python
standardized_data = standard_scaler.fit_transform(height_weight_age_df)
````

````python
standardized_df = pd.DataFrame(standardized_data, columns = ['Height', 'Weight', 'Age'])
standardized_df
````

````python
standardized_df.describe()
# 평균이 0 , 표준편차가 1
````

### One-hot Encoding

* pandas 에서 get_dummies() 메소드 사용하면 가능.
* 변환 원하는 dataframe 파리미터로.
* optional parmeter; column 지정해주면 데이터프레임에서 일부 컬럼만 변환 가능

````python
TITANIC_FILE_PATH = 'data/titanic.csv'
````

````python
titanic_df = pd.read_csv(TITANIC_FILE_PATH)
titanic_df.head()
````

````python
titanic_sex_embarked = titanic_df[['Sex', 'Embarked']]
titanic_sex_embarked
````

````python
one_hot_encoded_df = pd.get_dummies(titanic_sex_embarked) 
one_hot_encoded_df
````

원하는 열만 one hot encoding하기

````python
encoded_df = pd.get_dummies(data = titanic_df, columns = ['Sex', 'Embarked'])
encoded_df.head()
````
