박스 플롯


```python
exam_df = pd.read_csv('exam.csv')
exam_df
```

```python
exam_df['math score'].describe()
```

```python
exam_df.plot(kind = 'box', y = 'math score')
```

```python
exam_df.plot(kind = 'scatter', x = 'math score', y = 'writing score')
```

```python
exam_df.plot(kind = 'scatter', x = 'reading score', y = 'writing score')
```
